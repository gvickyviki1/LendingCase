# Data Cleaning

1. Remove completely null columns 
2. Remove columns having only one value in them

```python
completely_null = df.isnull().sum()[df.isnull().sum() == 39717].to_dict()
completely_null_key = list(completely_null.keys())
completely_null_key
```

```python
df.drop(columns =completely_null_key, inplace = True)
```

```python
df.isnull().sum()[(df.isnull().sum() > 0)]*100/(len(df))
```

```python
only_one_value = df.nunique()[df.nunique() == 1].to_dict()
only_one_value_columns = list(only_one_value.keys())
only_one_value_columns
```

```python
df.drop(columns =only_one_value_columns, inplace = True)
```

```python
df.isnull().sum()[(df.isnull().sum() > 0)]*100/(len(df))
```

```python
df.shape
```

```python
df['term'] = df['term'].apply(lambda x: int(x.strip().split(' ')[0]))
```

```python
df['int_rate'] = df['int_rate'].apply(lambda x: float(str(x).strip().split('%')[0]))
df['revol_util'] = df['revol_util'].apply(lambda x: float(str(x).strip().split('%')[0]))
df['emp_length'] = df['emp_length'].apply(lambda x: float(str(str(str(x).strip().split('year')[0]).split('+')[0]).strip().split('<')[-1].strip()))
```

```python
df1 = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
df1.shape
```

```python
df1.isnull().sum()[(df1.isnull().sum() > 0)]*100/(len(df1))
```

```python
df1.drop(columns=['next_pymnt_d','mths_since_last_delinq','mths_since_last_record'], inplace =True)
```

```python
df1.isnull().sum()[(df1.isnull().sum() > 0)]*100/(len(df1))
```

```python
df1['term'].value_counts()
```
